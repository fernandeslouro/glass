```
files_delete_dir = '/content/drive/MyDrive/Colab Notebooks/project-glass/images_to_delete'
to_delete = []
for file in os.listdir(files_delete_dir):
    with open(files_delete_dir + '/' + file) as f:
        content = f.readlines()
        content = [x.strip() for x in content] 
        to_delete += content
    print(len(to_delete))
to_delete.sort() 
print(to_delete)
```
